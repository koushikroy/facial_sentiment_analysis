```python
import cv2
import mediapipe as mp
import time
import os
import pandas as pd
import matplotlib.pyplot as plt
from shutil import copy

%matplotlib inline
```

## üê† Reading Image Path

```python
def list_files(dir):
    r = []
    for root, dirs, files in os.walk(dir):
        for name in files:
            r.append(os.path.join(root, name))
    return r
```

```python
root="./jaffedbase"
image_path=sorted(list_files(root))
print(len(image_path))

for image in image_path:
    print(image)
```

## ü¶É Define class names

```python
class_names = {
  "AN": "Angry",
  "DI": "Disgust",
  "FE": "Fear",
  "SA":"SAD",
  "SU":"Surprise",
  "HA":"Happy",
  "NE": "Neutral"
}
```

## ü¶î Create Directory for Each Class

```python
for emotion_label,emotion in class_names.items():
    os.mkdir(emotion)    
```

## üåé Classify Images 

```python
for image in image_path:
    for emotion_label,emotion in class_names.items():
        if emotion_label in image:
            copy(image,emotion)
          
```
